```python
ACCxval =[]
ACCyval =[]
ACCzval =[]
GYRxval =[]
GYRyval =[]
GYRzval =[]
MAGxval =[]
MAGyval =[]
MAGzval =[]
eAnglePval =[]  # Euler angles  Pitch
eAngleYval =[]  # Euler angles  Yaw
eAngleRval =[]  # Euler angles  Roll

for i in range(0,lenofIMUData):
    L1a = imuData[i].split()
    len1 = len(L1a)
    val1 = L1a[0]
    if val1 == '[[ACC':
        vaalX = float(L1a[2])
        ACCxval.append(vaalX)
        ACCyval.append(float(L1a[3])) 
        ACCzval.append(float(L1a[4])) 
    
    if val1 == '[[MAG' :
        MAGxval.append(float(L1a[2]))
        MAGyval.append(float(L1a[3])) 
        MAGzval.append(float(L1a[4])) 
    
    if val1 == '[[GYR' :
        GYRxval.append(float(L1a[2]))
        GYRyval.append(float(L1a[3]))
        GYRzval.append(float(L1a[4])) 
    
    if val1 == '[[COM' : # Euler angles
        eAnglePval.append(float(L1a[2]))
        eAngleYval.append(float(L1a[3]))
        eAngleRval.append(float(L1a[4])) 
```
